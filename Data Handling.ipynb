### Load useful Columns 

```python
cols_to_save = [
    'Created Date', 'Closed Date', 'Agency Name',
       'Problem (formerly Complaint Type)',
       'Problem Detail (formerly Descriptor)', 'Incident Zip',
       'Status', 'Due Date','Resolution Action Updated Date',
       'Police Precinct','Borough','Open Data Channel Type',
       'Latitude', 'Longitude','Street Name','Resolution Description'	
]

df = pd.read_csv('311_Service_Requests_Sample.csv' , usecols=cols_to_save , low_memory= False)
df
```

### Rename Columns

```python
'Open Data Channel Type'.lower().replace(' ','_')
```

```python
def rename_cols(x):
    return x.lower().replace(' ','_')

df.columns = df.columns.map(rename_cols)
df.columns
```

```python
df.rename(columns ={'problem_(formerly_complaint_type)' :'complaint_type',
                        'problem_detail_(formerly_descriptor)' : 'complaint_detail'} , inplace=True)
```

```python
df.columns
```

```python
df.info()
```

### Parsing Date from str to DateTime

```python
date_cols =['created_date' , 'closed_date' ,'due_date','resolution_action_updated_date']

df[date_cols] = df[date_cols].apply(pd.to_datetime, format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
```

### Check Missing Values

## NANs Handling Approach

- Keeping the NANs in [closed_date, resolution_action_updated_date] is meaningful —
  NaT indicates the complaint is still open or has no action taken yet.
- Replacing the NANs in [complaint_detail] with 'No Detail'.
- Replacing the NANs in [resolution_description] with 'No Resolution Yet'.
- Dropping the NANs in [street_name] and [incident_zip] as they represent
  a very small percentage of the data and borough already covers location.
- Dropping the NANs in both [latitude] and [longitude] 
- Dropping [due_date] column entirely as 99.7% of its values are missing.

```python
df['complaint_detail'] = df['complaint_detail'].fillna('No Detail')
```

```python
df['resolution_description'] = df['resolution_description'].fillna('No Resolution Yet')
```

```python
df.dropna(subset=['street_name' , 'incident_zip'] , inplace=True)
```

```python
df.dropna(subset=['latitude' , 'longitude'] , inplace=True) 
```

```python
df.drop(columns=['due_date'] , inplace=True)
```

```python
df.duplicated().sum()
```

```python
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)

print(f"Dropped duplicates: {before - after}")
print(f"Remaining rows: {after}")
```
